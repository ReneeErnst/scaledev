Visualize relationship between scale items

```python
import pandas as pd
import numpy as np
import seaborn as sns

import scaledev
from scaledev import preprocessor
```

```python
data_loc = scaledev.get_data_dir()

# Show full dataframes when printing
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", lambda x: "%.5f" % x)
np.set_printoptions(threshold=100, suppress=True)

# load raw data
df_raw = pd.read_excel(data_loc / "ets_data.xlsx")
df_raw.head()

df_raw = preprocessor.clean_columns(df_raw)
```

Examine relationships between vars as requested

```python
# Between cx9 and sp4
sns.lmplot(x="cx9", y="sp4", data=df_raw)
print("Correlation: ", df_raw["cx9"].corr(df_raw["sp4"]))
```

```python
# Between ae6 and to1
sns.lmplot(x="ae6", y="to1", data=df_raw)
print("Correlation: ", df_raw["ae6"].corr(df_raw["to1"]))
```

```python
# Between ae6 and to1
sns.lmplot(x="c5", y="c6", data=df_raw)
print("Correlation: ", df_raw["c5"].corr(df_raw["c6"]))
```
